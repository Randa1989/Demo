Handling the Null Values in the Dataset

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

Preprocessing the rating_count Column

```python
for i in df['rating_count'].values:
    print(i,'|',int(i.replace(',','')))
    break
```

```python
rating_count = []
for i in df['rating_count'].values:
    rating_count.append(int(i.replace(',','')))
```

```python
df['rating_count'] = rating_count
```

```python
df.head()
```

Preprocessing the discounted_price Column¶


```python
for i in df['discounted_price'].values:
    print(i,'|',float(i[1:].replace(',','')))
    break
```

```python
discounted_price = []
for i in df['discounted_price'].values:
    discounted_price.append(float(i[1:].replace(',','')))
```

```python
df['discounted_price'] = discounted_price
```

```python
df.head()
```

Preprocessing the actual_price Column¶


```python
for i in df['actual_price'].values:
    print(i,'|',float(i[1:].replace(',','')))
    break
```

```python
actual_price = []
for i in df['actual_price'].values:
    actual_price.append(float(i[1:].replace(',','')))
```

```python
df['actual_price'] = actual_price
```

```python
df.head()
```

Preprocessing the discount_percentage Column¶


```python
for i in df['discount_percentage'].values:
    print(i,'|',int(i[:-1])/100)
    break
```

```python
discount_percentage = []
for i in df['discount_percentage'].values:
    discount_percentage.append(int(i[:-1])/100)
```

```python
df['discount_percentage'] = discount_percentage
```

```python
df.head()
```

**Exploratory Data Analysis¶

Finding the Unique Category¶
**

```python
category = []
for i in df['category'].values:
    category += i.split('|')
```

```python
category = list(set(category))
```

```python
category
```

```python
len (category)
```

Finding the Products Distribution in each Category¶


```python
data = []
for i in category:
    c = 0
    for j in df.values:
        if(i in j[2]):
            c += 1
    data.append([i,c])
```

```python
data
```

**Sorting the Products Distribution in each Category based of the Frequency¶
**

```python
df_pr = pd.DataFrame(data,columns=['category','freq'])
```

```python
df_pr.sort_values(by='freq',ascending=False)
```
